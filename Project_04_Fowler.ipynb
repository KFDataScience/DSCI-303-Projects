## Part 1: Loading the Dataset; Preliminary Analysis

Here we will be importing the dataset and exploring the structure of the data.

```python
diamonds = pd.read_csv('diamonds.txt', sep='\t')

diamonds.head(10)
```

Next we will inspect the distribution of the columns in diamonds.

## Part 2: Filtering and Sorting

Now we will use filtering and sorting techniques to display info about the 5 most expensive diamonds in the diamonds dataset.

```python
diamonds_1 = diamonds[['price', 'carat', 'cut', 'color', 'clarity']]

diamonds_1['price'] = pd.to_numeric(diamonds_1['price'], errors='coerce')

diamonds_1_sorted = diamonds_1.sort_values(by='price', ascending=False)

diamonds_1_sorted.head(5)
```

Now we will explore info about the 5 least expensive diamonds in the dataset.

```python
diamonds_1_sorted1 = diamonds_1.sort_values(by='price', ascending=True)

diamonds_1_sorted1.head(5)
```

Next, we will explore the 5 largest diamonds in the dataset with an ideal cut.

```python
diamonds_1_sorted2 = diamonds_1_sorted1[diamonds_1_sorted1['cut'] == 'Ideal']

diamonds_1_sorted2 = diamonds_1_sorted2.sort_values(by='carat', ascending=False)

diamonds_1_sorted2.head()
```

Now we will view info about the 5 largest diamonds in the dataset with a fair cut.

```python
diamonds_1_sorted3 = diamonds_1_sorted1[diamonds_1_sorted1['cut'] == 'Fair']

diamonds_1_sorted3 = diamonds_1_sorted3.sort_values(by='carat', ascending=False)

diamonds_1_sorted3.head()
```

## Part 3: Working with Categorical Variables

Now we will be creating lists to specify the order for each of the three categorical variables.

```python
clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

cut_levels = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']

color_levels = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
```

Now we will use these lists to communicate to Pandas the correct order for the levels of the three cateogrical variables.

```python
diamonds['cut'] = pd.Categorical(diamonds['cut'], cut_levels)
diamonds['color'] = pd.Categorical(diamonds['color'], color_levels)
diamonds['clarity'] = pd.Categorical(diamonds['clarity'], clarity_levels)
```

Now we will create lists of named colors to serve as palettes to be used for visualizations later in the notebook.

```python
clarity_pal = ['red', 'blue', 'green', 'purple', 'orange', 'pink', 'brown', 'gray']
color_pal = ['red', 'blue', 'green', 'purple', 'orange', 'pink', 'brown']
cut_pal = ['red', 'blue', 'green', 'purple', 'orange']
```

## Part 4: Displaying Counts for Categorical Variables

Now we will determine the number of diamonds with each level of each of the three categorical variables.

We will start by counting the number of diamonds for each level of cut.

```python
cut_counts = diamonds['cut'].value_counts().sort_index()
cut_counts
```

Now we will count the number of diamonds for each level of color.

```python
color_counts = diamonds['color'].value_counts().sort_index()
color_counts
```

Now we will count the number of diamonds for each level of clarity.

```python
clarity_counts = diamonds['clarity'].value_counts().sort_index()
clarity_counts
```

## Part 5: Scatterplots of Price Against Carat

Now we will explore the relationship between the price and carat attributes by generating scatter plots.

```python
plt.figure(figsize=[8,6])
plt.scatter(diamonds['carat'], diamonds['price'], s=20, alpha=0.4, c='red', edgecolors='none')
plt.xlabel("Carat Size")
plt.ylabel("Price (in $)")
plt.title("Relationship between Diamond Price and Carat Size")
plt.show()
```

Now we will reproduce the plot from above, but with colored points according to their clarity.

```python
plt.figure(figsize=[8,6])

for i, level in enumerate(clarity_levels):
    plt.scatter(diamonds[diamonds['clarity'] == level]['carat'], 
                diamonds[diamonds['clarity'] == level]['price'], 
                s=20, alpha=0.4, c=clarity_pal[i], edgecolors='none', label=level)

plt.xlabel("Carat Size")
plt.ylabel("Price (in $)")
plt.title("Relationship between Diamond Price and Carat Size (Colored by Clarity)")
plt.legend(title="Clarity")
plt.show()
```

Now I will separate the points for each of the eight clarity levels into its own subplot.

```python
plt.figure(figsize=[12, 6])

for i, level in enumerate(clarity_levels):
    plt.subplot(2, 4, i+1)
    plt.scatter(diamonds[diamonds['clarity'] == level]['carat'], 
               diamonds[diamonds['clarity'] == level]['price'], 
               s=20, alpha=0.4, c=clarity_pal[i], edgecolors='none')
    plt.xlabel("Carat Size")
    plt.ylabel("Price (in $)")
    plt.title(f"Clarity: {level}")
    plt.xlim([0, 4])
    plt.ylim([0, 20000])

plt.tight_layout()
plt.show()
```

## Part 6: Applying Logarithmic Transformations

I will start by adding two new columns to the diamonds dataframe in order to store the transformed variables. I will then apply logarithmic transformations to the price and carat coloumns and explore the distribution of these transformed variables.

```python
diamonds['ln_carat'] = np.log(diamonds['carat'])
diamonds['ln_price'] = np.log(diamonds['price'])
```

I will now use histograms to explore the distribution of the diamond prices and the log of the diamond prices.

```python
plt.figure(figsize=[12,4])

# Histogram for Price
plt.subplot(1, 2, 1)
plt.hist(diamonds['price'], bins=20, edgecolor='black', color='blue')
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Prices")

# Histogram for Log Price
plt.subplot(1, 2, 2)
plt.hist(diamonds['ln_price'], bins=20, edgecolor='black', color='red')
plt.xlabel("Natural Log of Price")
plt.ylabel("Count")
plt.title("Histogram of Log Prices")

plt.tight_layout()
plt.show()
```

Now I will use histograms to explore the distributions of the carat sizes, and the log of the carat sizes.

```python
plt.figure(figsize=[12,4])

# Histogram for Carat Size
plt.subplot(1, 2, 1)
plt.hist(diamonds['carat'], bins=20, edgecolor='black', color='orange')
plt.xlabel("Carat Size")
plt.ylabel("Count")
plt.title("Histogram of Carat Sizes")

# Histogram for Log Carat Size
plt.subplot(1, 2, 2)
plt.hist(diamonds['ln_carat'], bins=20, edgecolor='black', color='green')
plt.xlabel("Natural Log of Carat Size")
plt.ylabel("Count")
plt.title("Histogram of Log-Carat Sizes")

plt.tight_layout()
plt.show()
```

## Part 7: Scatterplot of Transformed Variables

In this part, I will recreate the second scatterplot from part 5, but using the transformed variables instead of the original values. 

```python
plt.figure(figsize=[8,6])

for i, level in enumerate(clarity_levels):
    plt.scatter(diamonds[diamonds['clarity'] == level]['ln_carat'], 
                diamonds[diamonds['clarity'] == level]['ln_price'], 
                s=20, alpha=0.4, c=clarity_pal[i], edgecolors='none', label=level)

plt.xlabel("Natural Log of Carat Size")
plt.ylabel("Natural Log of Price")
plt.title("Relationship between Log-Price and Log-Carat Size")
plt.legend(title="Clarity")
plt.show()
```

## Part 8: Grouping by Cut

Now I will calculate the mean price and carat size for each level of the variable cut. I will create a grouped DataFrame displaying the mean price and carat size for each cut level.

```python
gb_cut = diamonds[['cut', 'price', 'carat']].groupby('cut').mean()

gb_cut
```

Now I will use bar charts to graphically display the info from the DataFrame above.

```python
plt.figure(figsize=[12,4])

plt.subplot(1, 2, 1)
plt.bar(gb_cut.index, gb_cut['price'], color=cut_pal, edgecolor='black')
plt.xlabel("Cut")
plt.ylabel("Mean Price")
plt.title("Mean Price by Cut")

plt.subplot(1, 2, 2)
plt.bar(gb_cut.index, gb_cut['carat'], color=cut_pal, edgecolor='black')
plt.xlabel("Cut")
plt.ylabel("Mean Carat Size")
plt.title("Mean Carat Size by Cut")

plt.tight_layout()
plt.show()
```

## Part 9: Grouping by Color

Now I will calculate the mean price and carat size for each level of the variable color. I will create a grouped DataFrame displaying the mean price and carat size for each color level.

```python
gb_color = diamonds[['color', 'price', 'carat']].groupby('color').mean()

gb_color
```

```python
plt.figure(figsize=[12,4])

plt.subplot(1, 2, 1)
plt.bar(gb_color.index, gb_color['price'], color=color_pal, edgecolor='black')
plt.xlabel("Color")
plt.ylabel("Mean Price")
plt.title("Mean Price by Color")

plt.subplot(1, 2, 2)
plt.bar(gb_color.index, gb_color['carat'], color=color_pal, edgecolor='black')
plt.xlabel("Color")
plt.ylabel("Mean Carat Size")
plt.title("Mean Carat Size by Color")

plt.tight_layout()
plt.show()
```

## Part 10: Grouping by Clarity

Now I will calculate the mean price and carat size for each level of the variable clarity. I will create a grouped DataFrame displaying the mean price and carat size for each level of clarity.

```python
gb_clarity = diamonds[['clarity', 'price', 'carat']].groupby('clarity').mean()

gb_clarity
```

Now I will use bar charts to graphically display the info from the DataFrame above.

```python
plt.figure(figsize=[12,4])

plt.subplot(1, 2, 1)
plt.bar(gb_clarity.index, gb_clarity['price'], color=clarity_pal, edgecolor='black')
plt.xlabel("Clarity")
plt.ylabel("Mean Price")
plt.title("Mean Price by Clarity")

plt.subplot(1, 2, 2)
plt.bar(gb_clarity.index, gb_clarity['carat'], color=clarity_pal, edgecolor='black')
plt.xlabel("Clarity")
plt.ylabel("Mean Carat Size")
plt.title("Mean Carat Size by Clarity")

plt.tight_layout()
plt.show()
```

In parts 8-10, you should notice that for each of the three categorical variables , higher quality diamonds tend to have a lower average price. This is because higher quality diamonds tend to be smaller. As we all know, diamond size has a much stronger effect on the price than cut, color, or clarity.
